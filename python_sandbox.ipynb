```python
def mutate(lst):
    lst.append(4)

nums = [1, 2, 3]
mutate(nums)

print(nums)  # [1, 2, 3, 4] ✅ changed
```

```python
def rebind(lst):
    lst = [9, 9, 9]

nums = [1, 2, 3]
rebind(nums)

print(nums)  # [1, 2, 3] ❌ unchanged
```

Mutable (can be changed in place)
- list
- dict
- set
- custom objects

Immutable (cannot be mutated)
- int
- float
- str
- tuple

In Python, mutations affect the original object because function parameters reference the same object, so in-place changes are visible to all references, while reassignment only changes the local reference.
